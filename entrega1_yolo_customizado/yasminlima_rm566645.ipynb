```python
from ultralytics import YOLO
```

```python

import yaml

data = {
    'path': '/content/drive/MyDrive/fiap',
    'train': 'train/images',
    'val': 'valid/images',
    'test': 'test/images',
    'names': {
        0: 'faca',
        1: 'celular'
    }
}

with open('/content/data.yaml', 'w') as f:
    yaml.dump(data, f)

```

```python
model = YOLO('yolov8n.pt')
```

```python
model.train(
    data='/content/data.yaml',
    epochs=30,
    imgsz=640,
    project='runs/detect',
    name='train_30'
)
```

```python
from ultralytics import YOLO

model = YOLO("yolov8n.pt")

model.train(
    data="/content/drive/MyDrive/fiap/data.yaml",
    epochs=60,
    imgsz=640,
    project="/content/drive/MyDrive/fiap",
    name="train_60"
)
```

```python
final_path = '/content/runs/detect/runs/detect/train_30'

from IPython.display import Image, display

display(Image(f'{final_path}/results.png'))
display(Image(f'{final_path}/confusion_matrix.png'))
```

O treinamento do modelo YOLO apresentou evolução consistente ao longo das épocas, com redução progressiva das métricas de erro (loss), indicando que o modelo conseguiu aprender padrões relevantes do dataset. As métricas de desempenho alcançadas, como mAP50 (~0.65) e mAP50-95 (~0.45), demonstram que o modelo possui uma capacidade razoável de detecção dos objetos propostos.

Entretanto, a análise da matriz de confusão evidenciou um desempenho desigual entre as classes. A classe "faca" apresentou alta taxa de acerto, enquanto a classe "celular" apresentou maior número de erros, indicando maior dificuldade de identificação.

Esses erros estão relacionados principalmente à baixa quantidade de imagens utilizadas no treinamento e à falta de diversidade no dataset, o que limita a capacidade de generalização do modelo. Além disso, a classe "celular" possui maior complexidade visual e pode se confundir com outros objetos, tornando sua detecção mais desafiadora em comparação com a classe "faca".

De forma geral, o modelo demonstrou boa capacidade de aprendizado, porém ainda possui margem para melhorias. O aumento da quantidade de dados, maior variedade nas imagens (ângulos, iluminação e contextos) e ajustes no treinamento podem contribuir significativamente para um desempenho mais equilibrado e preciso entre as classes.
